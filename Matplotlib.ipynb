```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.DataFrame()
df["a"] = np.random.randint(10, 50, 20)
df["b"] = np.random.randint(10, 50, 20)
df.head()
```

```python
plt.plot(df.a)

# Axis Labelling 
plt.xlabel("Index Number")
plt.ylabel("Values")

# Title Block
plt.title("Monthly Sales for our product line A")
plt.show()
```

```python
plt.plot(df.b)

# Axis Labelling 
plt.xlabel("Index Number")
plt.ylabel("Values")

# Title Block
plt.title("Monthly Sales for our product line B")
plt.show()
```

```python
plt.plot(df.a, label="Product A")
plt.plot(df.b, label="Product B")

plt.legend()
plt.title("Monthly sales comparision across product A & B")
plt.xlabel("Index")
plt.ylabel("Sales")
plt.show()
```

```python
# Chnage the size of graph

plt.figure(figsize=(10, 5))
plt.plot(df.a, label="Product A")
plt.plot(df.b, label="Product B")

plt.legend()
plt.title("Monthly sales comparision across product A & B")
plt.xlabel("Index")
plt.ylabel("Sales")
# Save the graph in a PNG file format use savefig
plt.savefig("monthly_comparision_wt_a_b.png")
plt.show()
```
