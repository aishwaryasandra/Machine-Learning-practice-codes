```python
# E commerce 
cust_table = pd.DataFrame(columns=["CustID", "CustName", "Address", "PhoneNumber"])
cust_table["CustID"] = [1,2,3, 4]
cust_table["CustName"] = ["ABC", "XYZ", "CDF", "GHI"]
cust_table["Address"] = ["St 34, CL", "St 45, TX", "St 78, NY", "St 91, WS"]
cust_table["PhoneNumber"] = [1234,7654, 234567, 6545]
```

```python
cust_table
```

```python
transaction_table = pd.DataFrame(columns=["CustID", "ProductID", "TotalAmount"])
transaction_table["CustID"] = [3, 3, 3, 1, 2, 3]
transaction_table["ProductID"] = ["3W", "3W", "2G", "2G", "5H", "5H"]
transaction_table['TotalAmount'] = [123, 123, 43, 43 ,321, 321]
```

```python
transaction_table
```

```python
# Merge/Join!!! 
```

```python
pd.merge(transaction_table, cust_table, how="inner", left_on="CustID", right_on="CustID")
```

```python
pd.merge(transaction_table, cust_table, how="inner", left_on="CustID", right_on="CustID").iloc[:, [0,1,2,3,5]]
```
